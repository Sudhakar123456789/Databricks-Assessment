#Customer Analytics

```python
%sql
CREATE OR REPLACE VIEW dev.gold.customer_risk_category_count AS
select risk_category,count(risk_category) risk_category_cnt from dev.silver.silver_customers group by risk_category
```

```python
%sql
SELECT * FROM dev.gold.customer_risk_category_count
```

```python
%sql
CREATE OR REPLACE VIEW dev.gold.avg_annual_income_by_state AS
select state,avg(annual_income) avg_annual_income from dev.silver.silver_customers group by state
```

```python
%sql
SELECT * FROM dev.gold.avg_annual_income_by_state;
```

```python
%sql
CREATE OR REPLACE VIEW dev.gold.age_of_customers as
SELECT customer_id,date_diff(year,date_of_birth,current_date) FROM DEV.silver.silver_customers
```

```python
%sql
select * from dev.gold.age_of_customers
```

```python
%sql
CREATE OR REPLACE VIEW dev.gold.top_10_customers_trans_amount as 
SELECT customer_id,sum(amount) as Trans_amount FROM dev.silver.silver_customer_account_txn group by customer_id
order by 2 desc limit 10
```

```python
%sql
select * from dev.gold.top_10_customers_trans_amount
```

#Account Analytics

```python
%sql
CREATE OR REPLACE VIEW dev.gold.active_vs_dormant_accounts as 
select * from dev.silver.silver_accounts where account_status in ('Dormant','Active')
```

```python
%sql
select * from dev.gold.active_vs_dormant_accounts
```

```python
%sql
CREATE OR REPLACE VIEW dev.gold.avg_balance_by_accounttype as
select account_type,avg(available_balance) from dev.silver.silver_accounts group by account_type
```

```python
%sql
select * from dev.gold.avg_balance_by_accounttype
```
